```python
sns.countplot(x='smoker', data=tips)
```

```python
sns.distplot(tips['total_bill'], bins=20)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.pairplot(tips, palette='rainbow', hue='size')
```

```python
sns.barplot(x='sex', y='tip', data=tips)
```

```python
sns.countplot(x='sex', data=tips, palette='Set3')
```

```python
sns.barplot(x='size', y='tip', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

```python
pivot_table = tips.pivot_table(values='total_bill', index='day', columns='sex')
```

```python
sns.heatmap(pivot_table)
```

```python
grid = sns.PairGrid(tips, hue='sex')
grid.map(plt.scatter)
plt.legend(loc=4)
```
